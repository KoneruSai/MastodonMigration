```python
import plotly.graph_objects as go

months = ['2022-11', '2022-12', '2023-01', '2023-02', '2023-03', '2023-04', '2023-05', '2023-06', '2023-07', '2023-08', '2023-09', '2023-10']
active_users = [7505, 5852, 4795, 3845, 3597, 3171, 2826, 2608, 3047, 2409, 2535, 2398]

user_changes = [active_users[0]] + [active_users[i+1] - active_users[i] for i in range(len(active_users) - 1)]


retention_percentage = [100] + [(active_users[i+1] / active_users[i]) * 100 for i in range(len(active_users) - 1)]

colors = {
    "increase": "#8FB9AB",
    "decrease": "#F08976",
    "total": "#314D63",
    "line": "#6B705C",
    "grid": "#B0B0B0"
}


fig = go.Figure()
fig.add_trace(go.Waterfall(
    x=months,
    measure=["absolute"] + ["relative"] * (len(months) - 2) + ["total"],
    y=user_changes,
    base=0,
    text=[f'{user_changes[i]}' for i in range(len(user_changes))],
    textposition="outside",
    decreasing={"marker": {"color": colors["decrease"]}},
    increasing={"marker": {"color": colors["increase"]}},
    totals={"marker": {"color": colors["total"]}},
    connector={"line": {"color": colors["grid"]}}
))


fig.add_trace(go.Scatter(
    x=months,
    y=retention_percentage,
    mode="lines+markers",
    line=dict(color=colors["line"], width=2),
    name="Retention Percentage",
    marker=dict(size=6, color=colors["line"]),
    yaxis="y2"
))


fig.update_layout(
    waterfallgroupgap=0.5,
    xaxis_title="Month",
    yaxis_title="Number of Users",
    yaxis2=dict(title="Monthly Retention Rate (%)", overlaying="y", side="right"),
    plot_bgcolor="white",
    showlegend=False,
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=True, gridcolor=colors["grid"]),

    margin=dict(l=50, r=50, t=50, b=50),
    width=1000,
    height=500
)

fig.show()
```
